```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import math
```

```python
# Reading the dataset
X = pd.read_csv("kmeansdata.csv")
X1 = X['x1']
X2 = X['x2']

# Normalizing the dataset
X1 = (X1-X1.mean())/X1.std()
X2 = (X2-X2.mean())/X2.std()

X = np.column_stack((X1, X2))
```

```python
# Plotting the normalized dataset
plt.figure(figsize=(6, 4))
    
plt.style.use('dark_background')
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(X1, X2)
plt.show()
```

```python
iters = 50
k = 2 # for 2 clusters
eps = 1e-4
np.random.seed(10)
C = np.random.random((k, 2))

final_cluster = []
for iteration in range(iters): 
    prev_C = C.copy()
    clusters = []
    
    for i in range(len(X)): 
        distances = []
        for j in range(k): 
            dist = np.sqrt((C[j][0] - X1[i])**2 + (C[j][1] - X2[i])**2) 
            distances.append(dist)
        clusters.append(np.argmin(distances))
    
    for i in range(k): 
        points = X[np.array(clusters) == i]
        C[i] = np.mean(points, axis = 0)

    diff = np.sum((C - prev_C)**2)
    if diff < eps:
        print(f"Converged after {iteration+1} iterations")
        final_cluster = clusters
        break
```

```python
# Plotting the clustered dataset 
plt.scatter(C[:, 0], C[:, 1], marker='x', color='g')

points = X[np.array(final_cluster) == 0] 
plt.scatter(points[:, 0], points[:, 1], color='b', label='Cluster 1')

points = X[np.array(clusters) == 1]
plt.scatter(points[:, 0], points[:, 1], color='r', label='Cluster 2')

plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Clustering (k = 2)")
plt.legend()
plt.show()
```
